```python
import pandas as pd
```

```python
df = pd.read_csv('Bike_Sales_2021.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# removing "$" and "," from columns
columns_to_clean = df[['Unit_Cost', 'Unit_Price',
       'Profit', 'Cost', 'Revenue']] 
for column in columns_to_clean:
    df[column] = df[column].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float) 

     
```

```python
df["Product"]= df["Product"].str.replace(",", " ", regex=False)
```

```python
df.head()
```

```python
df.to_csv('Bike_Sales_2021_cleaned.csv', index=False)
```
